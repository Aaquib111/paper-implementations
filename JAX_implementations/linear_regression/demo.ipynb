```python
import numpy as np
import jax.numpy as jnp
import matplotlib.pyplot as plt

from linear_regression import linear_regressor, update
```

# Demo Data

```python
N = 100 # number of samples
EPS = 0.01 # noise scale
SLOPE = 3 # slope of the line
INTERCEPT = 2 # intercept of the line

X = np.random.normal(size=(N,))
noise = np.random.normal(scale=EPS, size=(N,))
y = SLOPE * X + INTERCEPT + noise

# plot the data
plt.scatter(X, y)
plt.xlabel('X')
plt.ylabel('y')
plt.title(f'y = {SLOPE} * X + {INTERCEPT}')
plt.show()
```

```python
ITERATIONS = 1000
LEARNING_RATE = 0.01
params = jnp.array([1., 1.]) # initial parameters

for i in range(ITERATIONS):
    params = update(params, X, y, LEARNING_RATE)

# Plot the data and the fitted line
plt.scatter(X, y)
plt.plot(X, linear_regressor(params, X), color='red')
plt.xlabel('X')
plt.ylabel('y')
plt.title(f'y = {round(params[0], 3)} * X + {round(params[1], 3)}')
plt.show()
```
